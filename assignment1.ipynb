```python
#frequency of words 
word_count ={}

for word in cleaned_word_list:
    word = word.lower() #also counts the word if it also is at the beginning of a sentence
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print(word_count)
```

```python
#categorize the words

def categorize_word(word):
    nouns = ['ness', 'tion', 'ment', 'ity', 'er', 'or', 's', 'e', 'age']
    verbs = ['ate', 'ify', 'ize', 'en']
    adjectives = ['able', 'ible', 'ful', 'less', 'ous', 'ic', 'al', 'y']

    #checks if the word ends with any of the suffixes 
    for suffix in nouns:
        if word.endswith(suffix):
            return 'Noun'

    for suffix in verbs: 
        if word.endswith(suffix):
            return 'Verb'

    for suffix in adjectives:
        if word.endswith(suffix):
            return 'Adjective'

    return 'Unknown'

```

```python
def categorized_words(word_list):
    categorized_words_dict = {}
    for word in word_list:
        categorized_words_dict[word] = categorize_word(word)
    return categorized_words_dict
```

```python
categorized_words_dict = categorized_words(new_word_list)
print(categorized_words_dict)
```

```python
#combine all dictionries (count(frequency), length, word category)

combine_attributes_dict = {}

for key in word_count.keys():
    combine_attributes_dict[key] = [word_count[key], word_length[key], categorized_words_dict[key]]

print(combine_attributes_dict)

#OR 

#combine_attributes_dict2 = {key: [word_count[key], word_length[key], categorized_words_dict[key] for key in word_count.keys()}
```
